Refactor Lab 1 into a function that can be used over and over again

```python
import random

play_again = True
round_total = 0
round_current = 1
user_rolls = [] # History of user rolls
comp_rolls = [] # History of computer rolls
user_wins = [] # History of user round results (i.e. wins, losses, ties)

def game_init(): # Displays game rules
    print("Welcome to the Dice Game!",
          "Rules:",
          "\tYou choose the number of rounds against the computer.",
          "\tYou and the computer will each roll a dice.",
          "\tThe one with the higher number wins a point for the round.",
          "\tIf there is a tie, the player with the highest roll total wins.",
          "\tMatching rolls results in no points being rewarded.\n\n", sep='\n')

def game_reset(): # Resets variables for a new game
    global round_current, round_total, user_rolls, comp_rolls, user_wins
    round_total = 0
    round_current = 1
    user_rolls.clear()
    comp_rolls.clear()
    user_wins.clear()

def game_start(): # Asks for desired number of rounds
    global round_total
    while round_total < 5:
        try:
            round_total = int(input("Enter the number of rounds you would like to play (at least 5): "))
        except ValueError: #Ensures that an error is not returned when an integer is not inputted
            print("Please enter an integer greater than 5.")
    print('', f"{round_total} Rounds! Let's Go!", "Game Start!", sep='\n')

def round_start(): # Waits on user to begin the round
    global round_current
    print(f"Round {round_current}\n")
    input("Press Enter to roll the die...")

def roll_dice(): # Rolls the dice for the user and computer and adds the values to a history
    global user_rolls, comp_rolls, user_roll, comp_roll
    user_roll = random.randint(1, 6) # Generates number for user
    comp_roll = random.randint(1, 6) # Generates number for computer
    user_rolls.append(user_roll) # Adds dice roll to user history
    comp_rolls.append(comp_roll) # Adds dice roll to computer history
    print(f"You rolled: {user_roll}", 
    f"Computer rolled: {comp_roll}", sep='\n')

def print_round_result(): # Appends the result of the round to the user win history and prints it
    global user_roll, comp_roll, user_wins
    round_result_text = { # Dictionary of values that should be printed based on the result of the round
        "Win": "Congratulations! You won this round!",
        "Lose": "Darn! You lost. Better luck next round!",
        "Tie": "It's a tie!"
    }
    result = "Win" if user_roll > comp_roll else ( "Lose" if user_roll < comp_roll else "Tie" )
    user_wins.append(result)
    print( round_result_text[result] + "\n")

def round_end(): # Prepares variables for next round
    global round_current
    round_current += 1

def game_over(): # Lets player know that the rounds are over
    print("Game Over!", '', "Calculating Results...", sep='\n')
    input("Press any key to view results...")

def detect_tiebreaker(): # Detects if there was a tie and outputs the values needed to determine the final winner
    global round_total, user_rolls, comp_rolls, user_wins
    wins = user_wins.count("Win")
    losses = user_wins.count("Lose")
    if wins == losses:
        user_total = sum(user_rolls)
        comp_total = sum(comp_rolls)
        print("Woah! It was a tie!",
            "Calculating sum of rolls...", sep='\n\n')
        print(f"Your total rolls: {user_total}",
            f"Computer total rolls: {comp_total}", sep='\t')
    else:
        user_total = wins
        comp_total = losses
    return user_total, comp_total

def print_game_result(results): # Displays the results from the game
    wins, losses = results
    game_result_text = {
        "Win": "Congratulations! You won the game!",
        "Lose": "Darn! You lost. Better luck next time!",
        "Tie": "Woah! It was an absolute tie!"
    }
    result = "Win" if wins > losses else ("Lose" if losses > wins else "Tie")
    print(game_result_text[result])

def print_game_statistics(): # Displays game statistics
    global round_total, user_rolls, comp_rolls, user_wins
    print("\nDisplaying Statistics...")
    print("Round\t", "Your Rolls", "Comp Rolls","Round Results", sep='\t')
    
    for i in range(round_total):
        print(i+1, '', user_rolls[i], '', comp_rolls[i], '', user_wins[i], sep='\t')

def game_continue(): # Asks if the player would like to continue
    # Asks if the player would like to continue
    global play_again
    play_again = True if input("\nDo you want to play again? (y/n): ").lower() == "y" else False

def game_end(): # Closing statement
    print("\nThanks for playing!")

def main(): # Main game loop

    game_init()

    global play_again, round_total, round_current, user_rolls, comp_rolls, user_wins
    
    while play_again:

        game_reset()

        game_start()
        
        while round_current <= round_total: # Repeat until out of desired rounds
            
            round_start()

            roll_dice()

            print_round_result()

            round_end()

        game_over()
                
        print_game_result(detect_tiebreaker())

        print_game_statistics()

        game_continue()

    game_end()

if __name__ == "__main__":
    main()
```
